# Distribuciones sesgadas
Octavo día. Tras explorar la distribución normal y su utilidad para estimar lo “típico”, llegó el momento de analizar distribuciones que no tienen forma de campana simétrica. Así es, como se sugería en el cuaderno anterior, hay más tipos de distribuciones. Hoy analizaremos las distribuciones sesgadas, en las cuales los datos se concentran más hacia un extremo. Veremos cómo lucen, cómo se generan y cómo detectar su sesgo usando _Python_. Seguimos concentrados en distribuciones. Empecemos.

## ¿Qué es una distribución sesgada?
A diferencia de la distribución normal —que tiene su media y moda al centro y la famosa simetría de campana— las distribuciones sesgadas muestran una concentración de datos hacia uno de los lados:

- Sesgo a la derecha (positiva): la cola es más larga hacia valores altos (por ejemplo, ingresos o tiempos de espera).

- Sesgo a la izquierda (negativa): la cola es más larga hacia valores bajos (por ejemplo, edades de fallecimiento cuando la mayoría muere anciano).

Vamos a verlo con datos simulados.

## Simulación de una distribución sesgada
Como distribución sesgada nos iremos por una común como los ingresos. Ya sabes cómo funciona el mundo: muchas personas ganan poco y pocas ganan mucho.

```python
# Las tres mismas librerías del cuaderno de distribuciones normales.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Semilla para reproducibilidad.
np.random.seed(42)

# Simular datos con sesgo positivo (derecha).
ingresos = np.random.lognormal(mean=8, sigma=1, size=1000)
df = pd.DataFrame({'Ingreso': ingresos})

# Impresiones de comprobación.
print(f'Media: {df["Ingreso"].mean():.2f}')
print(f'Mediana: {df["Ingreso"].median():.2f}')
print(f'Desviación estándar: {df["Ingreso"].std():.2f}')
```

Presta atención a cómo la media y la mediana se alejan. Además la desviación estándar parece grande.

## Visualizando el sesgo
El histograma debería revelar el sesgo: la mayoría de los datos están cerca de valores bajos, pero hay una cola larga hacia la derecha. Se trata de las personas que ganan mucho dinero: una minoria. Maldito Elon Musk.

```python
plt.figure(figsize=(8,5))
plt.hist(df['Ingreso'], bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribución sesgada de ingresos')
plt.xlabel('Ingreso mensual')
plt.ylabel('Frecuencia')
plt.show()
```

## Análisis de la asimetría y más ejemplos de distribuciones sesgadas
Ya lo ves. Ésta es una distribución sesgada..., sesgada a la derecha. Te explicaré por qué. En distribuciones sesgadas:

- Cuando la media es mayor a la mediana tenemos un sesgo positivo (derecha).

- Cuando la media es menor a la mediana tenemos un sesgo negativo (izquierda).

Otros tipos de datos que suelen tener distribuciones sesgadas son la duración de vida útil de productos, precios de viviendas, tiempos de espera, duración de llamadas telefónicas, gastos médicos, ganacias en bolsa, etc. Ya entindes la idea.

## Conclusión
A diferencia de la distribución normal (analizada en el proyecto anterior), aquí la media no representa el “valor típico”; la mediana suele ser una mejor referencia central.
